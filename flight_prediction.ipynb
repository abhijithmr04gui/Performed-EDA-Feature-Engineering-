```python
dt['Date'] = dt['Date_of_Journey'].str.split('/').str[0]
dt['Month'] = dt['Date_of_Journey'].str.split('/').str[1]
dt['Year'] = dt['Date_of_Journey'].str.split('/').str[2]
```

```python
dt.head()
```

```python
dt['Date']  = dt['Date'].astype(int)
dt['Month'] = dt['Month'].astype(int)
dt['Year'] = dt['Year'].astype(int)
```

```python
dt.info()
```

```python
dt.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
dt['Arrival_Time']=dt['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
dt['Arrival_Hour']  = dt['Arrival_Time'].str.split(':').str[0]
dt['Arrival_Minute']= dt['Arrival_Time'].str.split(':').str[1]
```

```python
dt
```

```python
dt['Arrival_Hour'] = dt['Arrival_Hour'].astype(int)
dt['Arrival_Minute'] = dt['Arrival_Minute'].astype(int)
```

```python
dt.drop('Arrival_Time',axis = 1,inplace=True)
```

```python
dt['Dept_Hour']  = dt['Dep_Time'].str.split(':').str[0]
dt['Dept_Minute'] = dt['Dep_Time'].str.split(':').str[1]
```

```python
dt['Dept_Hour'] = dt['Dept_Hour'].astype(int)
dt['Dept_Minute'] = dt['Dept_Minute'].astype(int)
```

```python
dt.info()
```

```python
dt.drop('Dep_Time',inplace=True,axis=1)
```
